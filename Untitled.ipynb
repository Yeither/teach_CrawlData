```python
import urllib.request
import urllib.error
import http.client


def download_test(url, user_agent='yangyuye', num=2):
    print('downloading:', num, 'times', url)
    headers = {'user_agent': user_agent}
    request = urllib.request.Request(url, headers=headers)
    try:
        html = urllib.request.urlopen(request).read()
    except urllib.error.URLError as e:
        print('download error (URLError):', e.reason)
        html = None
        if num > 1:
            if hasattr(e, 'code') and 500 <= e.code < 600:
                return download_test(url, num=num - 1)
    except http.client.RemoteDisconnected as e:
        print('download error (RemoteDisconnected):', str(e))
        html = None
        if num > 1:
            return download_test(url, num=num - 1)
    return html


if __name__ == '__main__':
    # download_test('https://httpstat.us/500')
    download_test('https://httpbin.org/get')
```
