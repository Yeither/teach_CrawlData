## 处理POST请求

以百度翻译为例，当我输入"super"时，按f12,切换到网络监测，会发现一条sug信息请求方法为
POS，其负载的kw值为super

````python
import requests


url = "https://fanyi.baidu.com/sug"
data = {"kw": input("翻译：")}
resp = requests.post(url,data=data)
print(resp.json())                     # 百度翻译返回json文件
print('字符串：',resp.text)             # 用text会直接显示字符串
````

## re解析（正则表达式）

可以到一些[测试网站](https://tool.oschina.net/regex)自己试一下

```
.      匹配除换行以外字符
\w     匹配字母或数字或下划线
\s     匹配任意空白符
\d     匹配数字
\n     匹配一个换行符
\t     匹配一个制表符

^      匹配字符串的开始
$      匹配字符串的结尾

\W     匹配除了字母或数字或下划线
\D     匹配除了数字
\S     匹配除了空白符

a|b    匹配字符a或字符b
()     匹配括号内的表达式
[...]  匹配一个组里的字符，字符间不需要用逗号隔开
[^...] 匹配除了这个组里的字符，字符间不需要用逗号隔开

*      重复星号前面的字符大于等于零次都会匹配到
+      重复加号前面的字符大于等于一次都会匹配到
?      重复问号前面的字符零次或一次
{n}    重复{}前面的字符n次
{n,}   重复{}前面的字符n次或更多次
{n,m}  重复{}前面的字符n次到m次
```

```
.*     贪婪匹配
.*?    惰性匹配，有多个字段符合且有重复时，取最短的不重复的一个或数个字段
```

使用正则表达式提取数据后还可以对数据进行分类，通过
'''
(?P<种类名>正则表达式)
'''
来实现提取分类

````python
import re


text = ('''<div class='电话簿'><span id='12345678901'>张三</span></div>
         <div class='电话簿'><span id='22345678901'>李四</span></div>''')
obj = re.compile(r"<span id='(?P<id>\d+)'>(?P<name>.*?)</span>")
result = obj.finditer(text)

for item in result:
    id = item.group("id")
    print(id)
    name = item.group("name")
    print(name)
````

## re库实战，从豆瓣和电影网爬取电影数据

````python
import re
import requests


f = open("top250.csv", mode='w', encoding='utf-8')
url = "https://movie.douban.com/top250"
headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36 Edg/144.0.0.0"}

for i in range(0,11):
    params={'start':str((i-1)*25),'filter':''}
    resp = requests.get(url+'',headers=headers,params=params)
    resp.encoding='utf-8'
    content = resp.text
    #print(content)
    obj=re.compile( r'<div class="item">.*?<span class="title">(?P<name>.*?)</span>.*?'
                           r'导演: (?P<director>.*?)&nbsp.*?<br>'
                           r'\s+(?P<year>.*?)&nbsp.*?<span class="rating_num" property="v:average">'
                           r'(?P<point>.*?)</span>.*?<span>(?P<num>.*?)人评价</span>', re.S)
    result = obj.finditer(content)
    for item in result:
        print(item.group('name'))
        print(item.group('director'))
        print(item.group('year'), '年')
        print(item.group('point'), '分')
        print(item.group('num'), '人看过')
        print('-'*20)
        f.write(f"{item.group('name')},{item.group('director')},{item.group('year')},{item.group('point')},{item.group('num')}\n")

f.close()
resp.close()
````

xpath可以通过html的“地址”来定位数据，接下来演示取内容，各种子节点和id

````python
from lxml import etree


html = """  
        <html>
            <head>
              <meta charset="utf-8" />
                <title>页面上方标签</title>
            </head>
                <body bgcolor="pink">
                    <h1>h1是大标题</h1>
                    <p id='文本'>p是文本1</p>
                    <p id='文本'>p是文本2</p>
                    <a href="http://www.baidu.com">不知道的可以百度</a>
                </body>
                <p id='文本'>p是文本3</p>
        </html>

"""

et = etree.HTML(html)
result1 = et.xpath('/html/body/h1/text()')[0]
print(result1)
result2 = et.xpath('/html//p/text()')
print(result2)
result3 = et.xpath('/html/p/text()')
print(result3)
result4 = et.xpath('/html/*/p/text()')
print(result4)
result5 = et.xpath('/html/*/p/@id')
print(result5)
````

pyquery通过css选择器来选择数据

````python
from pyquery import PyQuery


html = """  
        <html>
            <head>
              <meta charset="utf-8" />
                <title>页面上方标签</title>
            </head>
                <body bgcolor="pink">
                    <h1>h1是大标题</h1>
                    <p id='文本'>p是文本1</p>
                    <p id='文本'>p是文本2</p>
                    <a href="http://www.baidu.com">不知道的可以百度</a>
                </body>
                <p id='文本'>p是文本3</p>
        </html>
"""

p = PyQuery(html)
passage = p("p")
print(passage.text())
passage1 = p("body")("p")
print(passage1.text())
passage3 = p("#文本")
print(passage3.text())
passage4 = p("[bgcolor='pink']")('a').attr("href")
print(passage4)
passage5 = p("p").items()
for item in passage5:
    print(item.text())
````

和别的不同的是pyquery还可以对html文件进行修改

````python
from pyquery import PyQuery


html = """  
        <html>
            <head>
              <meta charset="utf-8" />
                <title>页面上方标签</title>
            </head>
                <body bgcolor="pink">
                    <h1>h1是大标题</h1>
                    <p id='文本'>p是文本1</p>
                    <p id='文本'>p是文本2</p>
                    <a href="http://www.baidu.com">不知道的可以百度</a>
                </body>
                <p id='文本'>p是文本3</p>
        </html>
"""

p = PyQuery(html)
# after能在第一个符合条件的对象后添加
p("p#文本").after("""<div class="aaa">添加文本1</div>""")
print(p)
print("-"*20)
# append能在第一个符合条件的对象内部添加
p("p#文本").append("""<div class="aaa">添加文本2</div>""")
print(p)
print("-"*20)
# attr能在所有符合条件的对象内部添加属性,如果添加已有属性则覆盖
p("p#文本").attr("class","添加class")
print(p)
print("-"*20)
# remove_attr能在所有符合条件的对象内部删除属性
p("p#文本").remove_attr("class")
print(p)
print("-"*20)
# remove能删除所有符合条件的标签
p("p#文本").remove()
print(p)
print("-"*20)
````

```
<h1>标题1</h1>
<div class="container"> 文本1<div>

```
批量编写文本格式。也可以把<style>的类容单独写到一个.css文件中，但是要记得申明
```
<link rel="stylesheet" href="style.css">
```

# 多线程与多进程

## 多线程

````python
from threading import Thread

def func():
    for i in range(100):
        print("func",i)

if __name__ == '__main__':
    t=Thread(target=func)
    t.start()
    for i in range(100):
        print("main",i)
````

## 多进程

jupyter无法直接执行多进程任务，可以复制到IDE里用

````python
from multiprocessing import Process


def func():
    for i in range(1000):
        print("进程1--",i)

if __name__ == '__main__':
    p = Process(target=func)
    p.start()
    for j in range(1000):
       print("进程2--",j)
````
